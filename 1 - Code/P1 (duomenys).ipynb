## ANALIZE

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.stats.diagnostic import acorr_ljungbox
from statsmodels.tsa.seasonal import seasonal_decompose

# Load the provided dataset
file_path = 'SnaigeIS.csv'
df = pd.read_csv(file_path)

# Convert 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'])

# Set 'date' as index
df.set_index('date', inplace=True)

# Identify outliers in Revenue
Q1 = df['Revenue'].quantile(0.25)
Q3 = df['Revenue'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['Revenue'] < (Q1 - 1.5 * IQR)) | (df['Revenue'] > (Q3 + 1.5 * IQR))]

# Plotting Revenue and highlighting outliers
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Revenue'])
plt.scatter(outliers['Revenue'], [0]*len(outliers), color='red')
plt.title('Revenue Outliers')
plt.xlabel('Revenue')
plt.show()

# Plotting the histogram of Revenue
plt.figure(figsize=(10, 6))
sns.histplot(df['Revenue'], kde=True)
plt.title('Revenue Distribution')
plt.xlabel('Revenue')
plt.ylabel('Frequency')
plt.show()

# Shapiro-Wilk test for normality
shapiro_test = stats.shapiro(df['Revenue'])

# Augmented Dickey-Fuller test for stationarity
adf_test = adfuller(df['Revenue'])

adf_test_results = {
    'ADF Statistic': adf_test[0],
    'p-value': adf_test[1],
    'Critical Values': adf_test[4]
}

# Plotting Autocorrelation Function (ACF)
plt.figure(figsize=(10, 6))
plot_acf(df['Revenue'], lags=20)
plt.title('Autocorrelation of Revenue')
plt.show()

# Box-Pierce test for autocorrelation
box_pierce_test = acorr_ljungbox(df['Revenue'], lags=[20], return_df=True)

# Decomposing the Revenue series
decomposition = seasonal_decompose(df['Revenue'], model='additive', period=4)

# Plotting the decomposition
plt.figure(figsize=(12, 8))
decomposition.plot()
plt.show()

# Output the results
print("Shapiro-Wilk Test:", shapiro_test)
print("ADF Test Results:", adf_test_results)
print("Box-Pierce Test Results:", box_pierce_test)
```
